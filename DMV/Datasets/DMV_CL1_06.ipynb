```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"Datasets\retail_sales_dataset_corrected.csv")

# Explore the structure and content of the dataset
print(df.info())
print(df.describe())
print(df.head())
```

```python
df.columns
```

```python
region_sales = df.groupby('category')['price'].sum().reset_index()
region_sales = region_sales.sort_values(by='price', ascending=False)
print(region_sales)

top_regions = region_sales.head()
print(top_regions)
```

```python
import matplotlib.pyplot as plt

# Bar plot
plt.figure(figsize=(10,6))
plt.bar(region_sales['category'], region_sales['price'], color='skyblue')
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(8,8))
plt.pie(region_sales['price'], labels=region_sales['category'],
        autopct='%1.1f%%')
plt.title('Sales Distribution by Category')
plt.show()
```
